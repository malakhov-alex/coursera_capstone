```python
wiki_page = requests.get("https://en.wikipedia.org/wiki/List_of_neighborhoods_in_San_Francisco")
soup = BeautifulSoup(wiki_page.text, "lxml")

neighborhoods_list = []
stop_words = [
    'See also',
    'References',
    'External links',
    'Specific neighborhoods',
]
for tag in soup.find_all(class_="mw-headline"):
    if not tag.string in stop_words:
        neighborhoods_list.append(tag.string)
```

```python
for neighborhood in neighborhoods_list:
    address = {"Address" : "%s, San Francisco, CA" % neighborhood}
    neig = geocode(address, max_locations=1)
    print('Neighborhood: {}\t ShortLabel: {}\t\t Type: {}\t Nbrhd: {}' .format(neighborhood, neig[0]['attributes']['ShortLabel'], neig[0]['attributes']['Type'], neig[0]['attributes']['Nbrhd']))
    #print(neig[0]['attributes'])
```
